```python
# Define ideal and standard values
ideal = {
    'Temperature': 25,
    'Dissolved_Oxygen': 14.6,
    'pH': 7,
    'Conductivity': 1500,
    'BOD': 3,
    'Nitrate': 10,
    'Fecal_Coliform': 500,
    'Total_Coliform': 1000,
    'Fecal_Streptococci': 100
}

standard = {
    'Temperature': 40,
    'Dissolved_Oxygen': 5,
    'pH': 9,
    'Conductivity': 3000,
    'BOD': 6,
    'Nitrate': 45,
    'Fecal_Coliform': 2500,
    'Total_Coliform': 5000,
    'Fecal_Streptococci': 500
}

# Assign weights
weights = {
    'Temperature': 0.1,
    'Dissolved_Oxygen': 0.2,
    'pH': 0.15,
    'Conductivity': 0.05,
    'BOD': 0.1,
    'Nitrate': 0.1,
    'Fecal_Coliform': 0.1,
    'Total_Coliform': 0.1,
    'Fecal_Streptococci': 0.1
}
```

```python
import numpy as np
eps = 1e-9

# build a dataframe of q-values for each parameter
q_df = {}
for p, w in weights.items():
    avg_col = p + '_avg'
    if avg_col not in df.columns:
        raise KeyError(f"Expected column {avg_col} not found in df.columns")
    denom = (standard[p] - ideal[p]) if (standard[p] - ideal[p]) != 0 else eps
    q_df[p] = (df[avg_col].sub(ideal[p]).abs().div(denom).mul(100))

q_df = pd.DataFrame(q_df, index=df.index)

# compute weighted sum
weights_series = pd.Series(weights)
df['WQI'] = (q_df * weights_series).sum(axis=1)
```

```python
# define bins and labels as per your domain thresholds
bins = [0, 25, 50, 75, 100, np.inf]
labels = ["Excellent", "Good", "Moderate", "Poor", "Very Poor"]
df['WQI_cat'] = pd.cut(df['WQI'], bins=bins, labels=labels, right=False)
```
